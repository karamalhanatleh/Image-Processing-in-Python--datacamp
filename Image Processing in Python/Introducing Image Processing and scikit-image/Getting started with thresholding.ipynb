# Getting started with thresholding


```python
#import paskages
import matplotlib.pyplot as plt
from skimage.color import rgb2gray

from skimage.filters import threshold_otsu, threshold_local, threshold_multiotsu
from skimage import data
import numpy as np

from skimage.filters import try_all_threshold
```

```python
#visualng images
def show_image(image , title="Image" , cmap_type='gray'):
    plt.imshow(image , cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
#load image
image= plt.imread('../data/cam.jpeg')
```

```python
#optimal threshold
thresh=127
```

```python
# Apply thresholding
binary = image  >  thresh


#show original and thresolded
show_image(image,"Original")
show_image(binary[:,:,2],"Thresholded")
```

#### inverted thresholding


```python
#optimal threshold
thresh=127
# Apply thresholding
binary = image  <  thresh

#show original and thresolded
show_image(image,"Original")
show_image(binary[:,:,1],"Thresholded")
```

```python

# Load an example image
image = data.camera()

# Apply different thresholding algorithms
otsu_threshold = threshold_otsu(image)
local_threshold = threshold_local(image, block_size=35, offset=10)
multiotsu_thresholds = threshold_multiotsu(image, classes=4)

# Generate binary images using the thresholds
binary_otsu = image > otsu_threshold
binary_local = image > local_threshold
binary_multiotsu = np.digitize(image, bins=multiotsu_thresholds)

# Plot the original and thresholded images
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

axes[0, 0].imshow(image, cmap='gray')
axes[0, 0].set_title('Original Image')

axes[0, 1].imshow(binary_otsu, cmap='gray')
axes[0, 1].set_title('Otsu Thresholding')

axes[1, 0].imshow(binary_local, cmap='gray')
axes[1, 0].set_title('Local Thresholding')

axes[1, 1].imshow(binary_multiotsu, cmap='gray')
axes[1, 1].set_title('Multi-Otsu Thresholding')

for ax in axes.ravel():
    ax.axis('off')

plt.tight_layout()
plt.show()
```
