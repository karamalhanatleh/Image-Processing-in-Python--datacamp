```python
image = plt.imread('../data/chapter 1/fruits-2.jpg')
```

```python
plt.imshow(red,cmap='gray')
plt.title("Red")
plt.show()
```

```python
#visualng images
def show_image(image , title="Image" , cmap_type='gray'):
    plt.imshow(image , cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
##flip image in uo direction
vertically_flipped = np.flipud(madrid_imag)
show_image(madrid_imag, ' madrid_imag')
show_image(vertically_flipped, ' vertically flipped')
```
